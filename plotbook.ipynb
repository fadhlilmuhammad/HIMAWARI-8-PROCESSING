```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
folder = "/g/data/v46/fm6730/H8_TUTORIAL/L1_AND_L2/DATA/"
file   = "CTT_geocatL2.HIMAWARI-8.2022058.000000.FLDK.R20.nc"
```

```python
ds = xr.open_dataset(folder+file)
ds
```

```python
#Choose the variables

x = ds["Band1"]
x
```

```python
# Make figure and set the size
fig = plt.figure(figsize=(11,8.5))

# Set the map projection
ax=plt.axes(projection=ccrs.PlateCarree())

# Make shaded plot
cs = ax.contourf(x['lon'], x['lat'], x,
                 transform = ccrs.PlateCarree(),
                 cmap='YlOrBr',
                 extend='both')

# Add colorbar
cbar = plt.colorbar(cs,
                    shrink=0.7,
                    orientation='vertical',
                    label='[K]')

# Add coastlines
ax.coastlines()

```

```python
#save fig
plt.savefig("H8-CTT.png")
```
